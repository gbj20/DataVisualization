```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
	return 1 / (1 + np.exp(-x))

def relu(x):
	return np.maximum(0, x)

def tanh(x):
	return np.tanh(x)

def leaky_relu(x, alpha=0.01):
	return np.where(x > 0, x, x * alpha)

def elu(x, alpha=1.0):
	return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

def softmax(x):
	e_x = np.exp(x - np.max(x))
	return e_x / e_x.sum(axis=0)
```

```python
x = np.linspace(-5, 5, 100)
```

```python
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
y_leaky_relu = leaky_relu(x)
y_elu = elu(x)
y_softmax = softmax(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Functions')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y_relu, label='ReLU')

plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Functions')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y_tanh, label='Tanh')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Functions')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y_elu, label='ELU')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Functions')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y_softmax, label='Softmax')

plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Activation Functions')
plt.legend()
plt.show()
```
